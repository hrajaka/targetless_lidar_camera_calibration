```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread
from skimage import color

%matplotlib inline 
```

```python
def low_rank_approximation(X, rank):
    # low rank approximation
    SVD = np.linalg.svd(X, full_matrices=False)
    u,s,v = SVD
    res = np.zeros((375,1242))
    for i in range(rank):
        res += s[i] * np.outer(u.T[i],v[i])
    return res
```

```python
def mse(img1, img2):
    # python version of mean sqaure error
    err = np.sum((img1.astype("float") - img2.astype("float")) ** 2)
    err /= float(img1.shape[0] * img2.shape[1])
    return err
```

```python
road = imread("./training/image_2/000143.png")
road = color.rgb2gray(road)
plt.figure(figsize=(10, 5))
i0 = plt.subplot(221)
i0.set_title("original")
plt.imshow(road,cmap='gray')
road15 = low_rank_approximation(road, 15)
i1 = plt.subplot(222)
i1.set_title("rank-15")
plt.imshow(road15,cmap='gray')
road10 = low_rank_approximation(road, 10)
i2 = plt.subplot(223)
i2.set_title("rank-10")
plt.imshow(road10,cmap='gray')
road5 = low_rank_approximation(road, 5)
i3 = plt.subplot(224)
i3.set_title("rank-5")
plt.imshow(road5,cmap='gray')
plt.show()
print('mse of rank 15, 10 and 5')
print(mse(road15,road))
print(mse(road10,road))
print(mse(road5,road))
```
